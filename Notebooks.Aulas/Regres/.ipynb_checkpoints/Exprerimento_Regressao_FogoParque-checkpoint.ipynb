## Funções utilizadas

```python
#Função de leitura dos dados------------------------------------
def read_data(path):
    with open(path, newline='') as f:
        reader = csv.reader(f,delimiter=',')
        for row in reader:
            try:
                data = np.vstack( (data , np.asarray(row).astype(np.float) ) )
            except:
                data = np.asarray(row).astype(np.float)
    f.close()
    return data
```

## "Forest Fires Data Set" (http://archive.ics.uci.edu/ml/datasets/Forest+Fires)

#### Atributos:
    * Eixo x da coordenada espacial sobre o mapa do Paque Montesinho: 1 a 9
    * Eixo x da coordenada espacial sobre o mapa do Paque Montesinho: 1 a 9
    * Mês do ano: 1 a 12
    * Dia da semana: 1 a 7
    * Índice FFMC obtido do sistema FWI: 18.7 a 96.20
    * Índice DMC obtido do sistema FWI: 1.1 a 291.3
    * Índice DC obtido do sistema FWI: 7.9 a 860.6
    * Índice ISI obtido do sistema FWI: 0.0 a 56.10
    * Temperatura em garus Celsius: 2.2 a 33.30
    * Umidade relativa (%): 15.0 to 100
    * Velocidade do vento (km/h): 0.40 a 9.40
    * Chuva exterior (mm/m²): 0.0 a 6.4
    * Área de floresta queimada (ha): 0.00 to 1090.84

```python
path = 'forestfires_mod.csv'
nomes = ['CoordX','CoodY','Mês','DiaSem','FFMC','DMC','DC','ISI','TempC','UR','Vel.Vento','Chuva Ext.']

data = read_data(path)

x = data[:,0:12]
y = data[:,12]
```

```python
plt.scatter(x[:,0],x[:,1],s=y,c='r',alpha=0.20)
plt.xlabel('coord-X',fontsize=20)
plt.ylabel('coord-Y',fontsize=20)
```

```python
FS = (12,20) #Tamanho da figura a ser gerada
fig = plt.figure(constrained_layout=True,figsize=FS)
spec = gridspec.GridSpec(ncols=2, nrows=5, figure=fig)

atributos = [2,4,5,6,7,8,9,10,11]
pos = 0
for at in atributos:
    ax = fig.add_subplot(spec[ np.int64(pos/2),np.int64(pos%2)])
    ax.plot(x[:,at],y,'bo')
    ax.set_xlabel(nomes[at],fontsize=20)
    ax.set_ylabel('Área',fontsize=20)
    pos += 1
    
plt.show()
```

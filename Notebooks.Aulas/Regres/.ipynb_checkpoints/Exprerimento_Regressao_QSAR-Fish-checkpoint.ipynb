## Bibliotecas/módulos

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import csv

from sklearn import preprocessing
from sklearn.model_selection import train_test_split


from matplotlib import rcParams
rcParams['text.usetex'] = True
rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
```

## Funções utilizadas

```python
#Função de leitura dos dados------------------------------------
def read_data(path):
    with open(path, newline='') as f:
        reader = csv.reader(f,delimiter=',')
        for row in reader:
            try:
                data = np.vstack( (data , np.asarray(row).astype(np.float) ) )
            except:
                data = np.asarray(row).astype(np.float)
    f.close()
    return data
```

## "QSAR fish toxicity Data Set" (http://archive.ics.uci.edu/ml/datasets/QSAR+fish+toxicity)

#### M. Cassotti, D. Ballabio, R. Todeschini, V. Consonni. A similarity-based QSAR model for predicting acute toxicity towards the fathead minnow (Pimephales promelas), SAR and QSAR in Environmental Research (2015), 26, 217-243; doi: 10.1080/1062936X.2015.1018938

#### Atributos - 6 molecular descriptors and 1 quantitative experimental response:
* CIC0
* SM1_Dz(Z)
* GATS1i
* NdsCH
* NdssC
* MLOGP
* Resposta LC50 [-LOG(mol/L)]

```python
path = 'qsar_fish_toxicity.csv'
nomes = ['CICO','SM1','GATS','NdsCH','NdssC','MLOGP', 'Resp LC50']

data = read_data(path)

x = data[:,0:6]
y = data[:,6].reshape(-1,1) -1

#Ajuste e avaliação do modelo
percentualTeste = 0.33
xD, xI, yD, yI = train_test_split(x,y,test_size=percentualTeste)
```

```python
FS = (12,20) #Tamanho da figura a ser gerada
fig = plt.figure(constrained_layout=True,figsize=FS)
spec = gridspec.GridSpec(ncols=2, nrows=3, figure=fig)

atributos = [0,1,2,3,4,5]
atributoAlvo = 6
pos = 0
for at in atributos:
    ax = fig.add_subplot(spec[ np.int64(pos/2),np.int64(pos%2)])
    ax.plot(x[:,at],y,'bo')
    ax.set_xlabel(nomes[at],fontsize=20)
    ax.set_ylabel(nomes[atributoAlvo],fontsize=20)
    pos += 1
    
plt.show()
```

## Regressão linear com todas variáveis preditora

```python
#Instanciação do modelo
regLin = linear_model.LinearRegression()

#Ajuste do modelo
regLin.fit(xD,yD)

print(regLin.coef_)
print(regLin.intercept_)

#Avaliação do modelo
yEst = regLin.predict(xI)

mediaDesvios = np.sum( (yI - yEst)**2 )/len(yI)
print('Média dos desvios quadráticos: ', mediaDesvios)
```

## Regressão polinomial com todas variáveis preditora

```python
pol = PolynomialFeatures(1)
polXD = pol.fit_transform(xD)
polXI = pol.fit_transform(xI)

#Instanciação do modelo
regLin = linear_model.LinearRegression()

#Ajuste do modelo
regLin.fit(polXD,yD)

print(regLin.coef_)
print(regLin.intercept_)

#Avaliação do modelo
yEst = regLin.predict(polXI)

mediaDesvios = np.sum( (yI - yEst)**2 )/len(yI)
print('Média dos desvios quadráticos: ', mediaDesvios)
```

## Visualização de regressão polinomial (grau 4) com 2 variáveis preditoras

```python
parSelecionado = [2,5]
pol = PolynomialFeatures(4)
polXD = pol.fit_transform(xD[:,parSelecionado])

#Instanciação do modelo
regLin = linear_model.LinearRegression()

#Ajuste do modelo
regLin.fit(polXD,yD)

print(regLin.coef_)
print(regLin.intercept_)


%matplotlib notebook

#Modelo de predição...
N = 100
X1 = np.linspace(np.min(x[:,parSelecionado[0]]),np.max(x[:,parSelecionado[0]]),N)
X2 = np.linspace(np.min(x[:,parSelecionado[1]]),np.max(x[:,parSelecionado[1]]),N)

mX1,mX2 = np.meshgrid(X1,X2)
mX12 = np.array([np.ravel(mX1),np.ravel(mX2)]).T

polX12 = pol.fit_transform(mX12)

yEst = regLin.predict(polX12)
yEst = yEst.reshape(mX1.shape)

fig = plt.figure(figsize=(7, 7))
elev = 45
azim = 45
ax = Axes3D(fig, elev=elev, azim=azim)

ax.scatter(x[:,parSelecionado[0]], x[:,parSelecionado[1]], y, c='r', marker='o')
ax.plot_surface(mX1, mX2, yEst, color='blue',alpha=0.5)
ax.set_xlabel(nomes[parSelecionado[0]],fontsize=20)
ax.set_ylabel(nomes[parSelecionado[1]],fontsize=20)
ax.set_zlabel(nomes[6],fontsize=20)
```
